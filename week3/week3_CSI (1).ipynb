```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")  # Clean style
```

```python
# Load the dataset
df = sns.load_dataset("tips")
df.head()
```

```python
# Basic information
df.info()
```

```python
# Summary statistics
df.describe()
```

```python
# Distribution of Total Bill
plt.figure(figsize=(8, 4))
sns.histplot(df['total_bill'], kde=True, color='skyblue')
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
# Distribution of Tip
plt.figure(figsize=(8, 4))
sns.histplot(df['tip'], kde=True, color='orange')
plt.title("Distribution of Tip Amount")
plt.xlabel("Tip")
plt.ylabel("Frequency")
plt.show()
```

```python
# Total Bill vs Tip
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="total_bill", y="tip", hue="sex", style="smoker")
plt.title("Tip vs Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

```python
# Boxplot of Tip by Day
plt.figure(figsize=(8, 6))
sns.boxplot(x="day", y="tip", data=df, palette="Set2")
plt.title("Tips by Day of the Week")
plt.xlabel("Day")
plt.ylabel("Tip Amount")
plt.show()
```

```python
# Violin plot of Tip by Time
plt.figure(figsize=(8, 6))
sns.violinplot(x="time", y="tip", data=df, hue="sex", split=True)
plt.title("Tip Distribution by Time of Day and Gender")
plt.xlabel("Time")
plt.ylabel("Tip")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Tip Percentage by Gender and Time
df["tip_pct"] = df["tip"] / df["total_bill"] * 100
plt.figure(figsize=(8, 6))
sns.barplot(x="sex", y="tip_pct", hue="time", data=df)
plt.title("Average Tip Percentage by Gender and Time")
plt.ylabel("Tip %")
plt.show()
```

# Conclusion and Analysis of Tips Dataset EDA

After conducting a detailed exploratory data analysis (EDA) on the Tips dataset, the following insights and patterns were identified:

1. Relationship Between Total Bill and Tip:
   There is a strong positive correlation between the total bill and tip amount—customers who spend more generally tip more. However, the increase in tips is not perfectly proportional, indicating varying tipping behavior at higher spending levels.

2. Gender and Tipping Behavior:
   Male customers tend to tip slightly more on average than females. The difference is more evident during dinner hours, suggesting gender-based behavioral trends in tipping.

3. Influence of Day and Time:
   Sunday and Saturday record the highest number of transactions and higher total bills. Dinner service yields larger bills and tips compared to lunch, showing peak business and generosity in the evening.

4. Tip Percentage (Tip as % of Total Bill):
   Tip percentages typically range from 10% to 20%, with minor variations based on gender, day, and time. Males tipping during dinner hours show a slightly higher average tip percentage.

5. Effect of Group Size:
   As the group size increases, the total bill increases, but tip percentage tends to decline slightly, indicating tip splitting or group-sharing effects.

6. Smoker vs Non-Smoker:
   There is no significant difference in tipping behavior between smokers and non-smokers, although non-smokers exhibit more consistent tipping patterns.

7. Distribution of Features:
   The majority of tips fall below $5, and most meals cost between $10–20, indicating a moderate spending pattern among the sampled restaurant customers.

8. Correlation Analysis:
   The highest correlation exists between total_bill and tip (~0.68), followed by a moderate correlation between size and total_bill. Other variables such as sex, time, or smoker are more useful for categorical comparisons rather than direct numerical correlation.


Final Thoughts:
The Tips dataset offers rich insights into customer tipping behavior across various social and service variables. Through visual analysis, it becomes evident how time, gender, amount spent, and party size influence tipping decisions. This dataset serves as a strong example of how simple real-world transactional data can reveal meaningful business and behavioral insights.
